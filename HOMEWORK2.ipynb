### Bài 3
 Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình.

```python
import math
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))
if (a + b > c) and (a + c > b) and (b + c > a):
    if a == b == c:
        loai_tg = "Tam giác đều"
    elif a == b or b == c or a == c:
        if (a**2 + b**2 == c**2) or (a**2 + c**2 == b**2) or (b**2 + c**2 == a**2):
            loai_tg = "Tam giác vuông cân"
        else:
            loai_tg = "Tam giác cân"
    elif (a**2 + b**2 == c**2) or (a**2 + c**2 == b**2) or (b**2 + c**2 == a**2):
        loai_tg = "Tam giác vuông"
    else:
        loai_tg = "Tam giác thường"
    chu_vi = a + b + c
    p = chu_vi / 2 
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"Đây là: {loai_tg}")
    print(f"Chu vi: {chu_vi}")
    print(f"Diện tích: {dien_tich:.2f}")

else:
    print("Khong la ba canh cua mot tam giac")
```

### Bài 8
 Giải hệ phương trình tuyến tính: 
ax  by  m 
cx  dy  n 

```python
print("Nhập a, b, m, c, d, n cách nhau bởi dấu cách:")
a, b, m, c, d, n = map(float, input().split())

# Tính các định thức (Công thức Cramer)
D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c
if D != 0:
    x = Dx / D
    y = Dy / D
    print(f"Hệ có nghiệm duy nhất: x = {x}, y = {y}")
else:
    if Dx == 0 and Dy == 0:
        print("Hệ có vô số nghiệm")
    else:
        print("Hệ vô nghiệm")
```

### Bài 9
 Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết 
ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao 
nhiêu giây. 


```python
t = float(input())
if t > 0 and t < 60:
    print(f"{t} giay")
elif t >= 60 and t < 3600:
    phut = t // 60
    giay = t % 60
    print(f"{phut} phut {giay} giay")
else: # Trường hợp t >= 3600
    a = t // 3600 # Giờ
    con_du = t % 3600
    b = con_du // 60 # Phút
    c = con_du % 60 # Giây
    print(f"{a} gio {b} phut {c} giay")
```

```python
t = int(float(input("Nhập số giây: "))) # Ép kiểu về int để tính toán chính xác

# Tính toán
gio = t // 3600
phut = (t % 3600) // 60
giay = t % 60

# In kết quả
print(f"{gio} giờ, {phut} phút, {giay} giây")
```

### Bài 10
Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một 
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
import math
x1, y1 = map(float, input("Nhập tọa độ tâm I (x y): ").split())
r = float(input("Nhập bán kính R: "))
x2, y2 = map(float, input("Nhập tọa độ điểm A (x y): ").split())
d = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
if d == r:
    print("Điểm A thuộc đường tròn.")
elif d < r:
    print("Điểm A nằm trong đường tròn.")
else:
    print("Điểm A nằm ngoài đường tròn.")
```
