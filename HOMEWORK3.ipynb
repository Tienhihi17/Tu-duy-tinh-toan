### Bài 4:
Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím


```python
# Nhập độ cao h từ bàn phím
h = int(input("Nhập độ cao h: "))

# Duyệt qua từng dòng từ 1 đến h
for i in range(1, h + 1):
    # 1. In khoảng trắng phía trước để tạo độ nghiêng cho tam giác
    # Số khoảng trắng ở mỗi dòng sẽ là h - i
    print(" " * (h - i), end="")
    
    # 2. Xử lý in dấu sao
    if i == 1:
        # Dòng đầu tiên chỉ có 1 dấu sao
        print("*")
    elif i == h:
        # Dòng cuối cùng in đầy đủ các dấu sao (tổng cộng 2*h - 1 dấu sao)
        print("*" * (2 * i - 1))
    else:
        # Các dòng ở giữa: in 1 dấu sao, khoảng trống ở giữa, và 1 dấu sao
        # Số khoảng trống ở giữa được tính bằng công thức: 2*i - 3
        print("*" + " " * (2 * i - 3) + "*")
```

```python
A = float(input("Nhập A: "))
s = 0
n = 0
while s <= A:
    n += 1
    s += 1/n
print(f"Số n nhỏ nhất là: {n} (với tổng S = {s:.4f} > {A})")
```

```python

n = int(input("Nhập số nguyên (nhập -1 để dừng): "))

if n == -1:
    print("Dãy rỗng.")
else:
    max_val = n
    min_val = n
    
    while True:
        n = int(input("Nhập số tiếp theo: "))
        if n == -1:
            break
        if n > max_val:
            max_val = n
        if n < min_val:
            min_val = n
            
    print(f"Số lớn nhất là: {max_val}")
    print(f"Số nhỏ nhất là: {min_val}")
```

```python
n = int(input("Nhập n: "))
temp = n
dem = 0
if temp == 0: dem = 1
while temp > 0:
    dem += 1
    temp //= 10
print(f"Số chữ số (C1): {dem}")
```

```python
n = int(input("Nhập n: "))
chan = 0
le = 0
temp = n
if temp == 0: chan = 1
while temp > 0:
    chu_so = temp % 10
    if chu_so % 2 == 0:
        chan += 1
    else:
        le += 1
    temp //= 10
print(f"Số chữ số chẵn: {chan}, số chữ số lẻ: {le}")
```

```python
n = int(input("Nhập n: "))
is_power_of_3 = True
if n <= 0:
    is_power_of_3 = False
else:
    temp = n
    while temp > 1:
        if temp % 3 != 0:
            is_power_of_3 = False
            break
        temp //= 3

if is_power_of_3:
    print(f"{n} có dạng 3^k")
else:
    print(f"{n} không có dạng 3^k")
```

```python
A = int(input("Nhập A: "))
f1, f2 = 1, 1
fib_max = 1

while f2 <= A:
    fib_max = f2
    f1, f2 = f2, f1 + f2

print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {fib_max}")
```

```python
n = int(input("Nhập n: "))
dem = 0
temp = n
while temp > 0:
    chu_so = temp % 10
    # Kiểm tra trực tiếp các số nguyên tố có 1 chữ số
    if chu_so in [2, 3, 5, 7]:
        dem += 1
    temp //= 10
print(f"Số lượng chữ số là số nguyên tố: {dem}")
```
