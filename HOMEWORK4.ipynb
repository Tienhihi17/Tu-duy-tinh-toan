```python
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        print(i, end=" ")
```

Nhập n
Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 


```python
n = int(input("Nhập n: "))
s = 0
k = 0
while s + (k + 1) < n:
    k += 1
    s += k
print(f"k lớn nhất là: {k} (Tổng S = {s})")
```

Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
day_so = []
while True:
    so = int(input("Nhập số (-1 để dừng): "))
    if so == -1:
        break
    day_so.append(so)

if day_so:
    print(f"Lớn nhất: {max(day_so)}, Nhỏ nhất: {min(day_so)}")
```

Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python
A = int(input("Nhập A: "))
f1, f2 = 1, 1
f_max = 1
while f2 <= A:
    f_max = f2
    f1, f2 = f2, f1 + f2
print(f"Số Fibonacci lớn nhất <= A là: {f_max}")
```

Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng)
Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
s = input("Nhập chuỗi: ")
count = 0
in_word = False
for i in range(1000000): # Giả lập vòng lặp qua index
    try:
        char = s[i]
        if char != " " and not in_word:
            count += 1
            in_word = True
        elif char == " ":
            in_word = False
    except IndexError:
        break
print("Số từ:", count)
```

Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành
thường và thường thành hoa. 

```python
s = input("Nhập chuỗi: ")
ket_qua = ""
for char in s:
    if 'A' <= char <= 'Z':
        ket_qua += chr(ord(char) + 32)
    elif 'a' <= char <= 'z':
        ket_qua += chr(ord(char) - 32)
    else:
        ket_qua += char
print(ket_qua)
```

Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính
tổng 3 số nguyên đó
VD:
Nhập: 3, 12, 15
Tổng: 30

```python
# Ví dụ nhập: 3, 12, 15
s = input("Nhập 3 số: ")
temp_num = ""
tong = 0
for char in s:
    if char != ",":
        temp_num += char
    else:
        tong += int(temp_num)
        temp_num = ""
tong += int(temp_num) # Cộng số cuối cùng
print("Tổng:", tong)
```

Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số 

```python
s = input("Nhập chuỗi: ")
hoa = thường = số = 0
for char in s:
    if 'A' <= char <= 'Z': hoa += 1
    elif 'a' <= char <= 'z': thường += 1
    elif '0' <= char <= '9': số += 1
print(f"Hoa: {hoa}, Thường: {thường}, Số: {số}")
```
