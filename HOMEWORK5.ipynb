Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_armstrong(x):
    temp = x
    n = 0
    # Đếm số chữ số
    while temp > 0:
        temp //= 10
        n += 1
    
    temp = x
    tong = 0
    while temp > 0:
        chu_so = temp % 10
        tong += chu_so ** n
        temp //= 10
    return tong == x
```

Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_perfect_square(x):
    if x < 0: return False
    i = 0
    while i * i <= x:
        if i * i == x:
            return True
        i += 1
    return False
```

 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.


```python
def is_palindrome(x):
    if x < 0: return False
    temp = x
    dao_nguoc = 0
    while temp > 0:
        dao_nguoc = dao_nguoc * 10 + (temp % 10)
        temp //= 10
    return dao_nguoc == x
```

Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.


```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except ValueError:
        try:
            float(s)
            return 'float'
        except ValueError:
            return 'neither'
```

 Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

```python
def is_one_product_of_others(a, b, c):
    return a == b * c or b == a * c or c == a * b
```

 Viết hàm sum_halves(s) nhận chuỗi 6 chữ số và trả về tổng của hai số 3 chữ số đầu và 3 chữ số cuối.

```python
def sum_halves(s):
    first_half = int(s[:3])
    second_half = int(s[3:])
    return first_half + second_half
```

Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
def quadratic_roots_count(a, b, c):
    if a == 0:
        return 1 if b != 0 else (0 if c != 0 else -1) # -1 đại diện vô số nghiệm
    delta = b**2 - 4*a*c
    if delta > 0: return 2
    elif delta == 0: return 1
    else: return 0
```

Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    # Sử dụng diện tích tam giác bằng 0
    return (x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)) == 0
```

Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def count_non_zero(lst):
    # Giả sử lst là list các số nguyên nhập từ bàn phím
    count = 0
    for num in lst:
        s = str(abs(num))
        for digit in s:
            if digit != '0':
                count += 1
    return count
```

Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].

```python
def get_max_min(a, b, c):
    lst = [a, b, c]
    return [max(lst), min(lst)]
```

Viết hàm tính hệ số nhị thức C(n, k).

```python
import math
def combinations(n, k):
    if k < 0 or k > n: return 0
    return math.comb(n, k)
```

Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,
giá km, phí chờ, phụ phí ban đêm và giảm giá.

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    # Đây là khung logic, giá trị cụ thể tùy thuộc quy định đề bài (không ghi rõ trong ảnh)
    price_per_km = 15000 if car_type == 4 else 17000
    fare = km * price_per_km + (wait_minutes * 1000)
    if 22 <= start_hour or start_hour <= 5: # Phụ phí đêm
        fare *= 1.1
    return fare
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n_unique_digits(n):
    if n <= 0: return 0
    tong = 0
    for i in range(1, n + 1):
        s = str(i)
        if len(s) == len(set(s)): # Nếu độ dài chuỗi bằng độ dài tập hợp các chữ số
            tong += i
    return tong
```
